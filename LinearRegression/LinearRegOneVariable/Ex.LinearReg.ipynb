1.Carregando os dados e adicionando em um array

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
X=[]
Y=[]

for linha in open("/content/drive/My Drive/Colab Notebooks/MachineLearning/ex1data1.txt"):
  x, y = linha.split(',') # Separa linhas por ","
  X.append(float(x))
  Y.append(float(y))

print(X,"\n",Y)
```
